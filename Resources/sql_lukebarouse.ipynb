```python
%%sql
Select
orderdate,
quantity,
netprice,
case
when quantity >=2 and netprice >= 50 THEN 'High Value Order'
else 'standard order'
END as order_type
from
sales
limit 10;
```

```python
%%sql

SELECT
orderdate,
quantity,
netprice,
CASE
WHEN quantity >= 2 AND netprice >= 100 THEN 'Multiple High Value Orders'
WHEN netprice >= 100 THEN 'Single HIGH Value Item'
WHEN quantity >=2 THEN 'Multiple Standard Items'
ELSE 'Single Standard Item'
END AS order_type
FROM sales
limit 10
```

```python
%%sql
SELECT p.categoryname,
     AVG(CASE WHEN s.orderdate BETWEEN'2022-01-01'AND '2022-12-31' THEN (s.quantity * s.netprice * s.exchangerate) END) AS total_avg_revenue_2022,
     AVG(CASE WHEN s.orderdate BETWEEN'2023-01-01'AND '2023-12-31' THEN (s.quantity * s.netprice * s.exchangerate) END) AS total_avg_revenue_2023
     FROM sales s LEFT JOIN product p ON s.productkey = p.productkey
     GROUP BY p.categoryname
     ORDER BY p.categoryname;
```

```python
%%sql
SELECT
PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY (s.quantity*s.netprice*s.exchangerate))AS median
FROM
sales s
where orderdate between '2022-01-01' AND '2023-12-31'
```

```python
%%sql

WITH median_value as (
SELECT
PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY (s.quantity*s.netprice*s.exchangerate))AS median
FROM
sales s
where orderdate between '2022-01-01' AND '2023-12-31'
)
SELECT
p.categoryname AS category,
SUM(CASE WHEN (s.quantity * s.netprice *s.exchangerate)< mv.median AND
  s.orderdate BETWEEN '2022-01-01' AND '2022-12-31'
 THEN(s.quantity * s.netprice *s.exchangerate) END) AS low_net_revenue_2022,
 SUM(CASE WHEN (s.quantity * s.netprice *s.exchangerate) >= mv.median AND
 s.orderdate BETWEEN '2022-01-01' AND '2022-12-31'
 THEN(s.quantity * s.netprice *s.exchangerate) END) AS high_net_revenue_2022,
SUM(CASE WHEN (s.quantity * s.netprice *s.exchangerate)< mv.median AND
 s.orderdate BETWEEN '2023-01-01' AND '2023-12-31'
 THEN(s.quantity * s.netprice *s.exchangerate) END) AS low_net_revenue_2023,
  SUM(CASE WHEN (s.quantity * s.netprice *s.exchangerate) >= mv.median AND
 s.orderdate BETWEEN '2023-01-01' AND '2023-12-31'
 THEN(s.quantity * s.netprice *s.exchangerate) END) AS high_net_revenue_2023
FROM
sales s
LEFT JOIN product p ON s.productkey = p.productkey,
median_value mv
GROUP BY
p.categoryname
ORDER BY
p.categoryname;
```

```python
%%sql
WITH percentiles AS(
SELECT
PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY (s.quantity*s.netprice*s.exchangerate)) AS revenue_25thpecentile,
PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY (s.quantity*s.netprice*s.exchangerate)) AS revenue_75thpecentile
from
sales s
where
orderdate BETWEEN '2022-01-01' AND '2023-12-31'
)
SELECT
p.categoryname AS category,
CASE  WHEN (s.quantity*s.netprice*s.exchangerate) <= pctl.revenue_25thpecentile Then '3-LOW'
      WHEN (s.quantity*s.netprice*s.exchangerate) >= pctl.revenue_75thpecentile Then '1- HIGH'
      ELSE '2-MEDIUM'
END AS revenue_tier,
sum(s.quantity*s.netprice*s.exchangerate)as total_revenue
FROM
sales s
LEFT JOIN product p ON s.productkey = p.productkey,
percentiles pctl
GROUP BY
p.categoryname,
revenue_tier
ORDER BY
p.categoryname,
revenue_tier
```
